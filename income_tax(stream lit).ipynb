```python
%%writefile adityaui.py
import streamlit as st
st.title("INCOME TAX CALCULATOR:")



income_1=st.number_input("Enter income from salary:")
income_2=st.number_input("Enter income from interest:")
income_3=st.number_input("Enter income from home loan self occupied:")
income_4=st.number_input("Enter income from rental:")
income_5=st.number_input("Enter income from home-loan let out:")
income_6=st.number_input("Enter income from digital assets:")
income_7=st.number_input("Enter income from others:")


income=income_1+income_2+income_3+income_4+income_5+income_6+income_7

st.write("Your total income is:",income)

regime=st.radio("regime:",["new",'old'])

standard=50000


if regime=="old":
    
    #if st.button("submit"):
        
    d1=st.number_input("enter 80C deduction max upto1.5lakh:")
    d2=st.number_input("enter 80TTA deduction:")
    d3=st.number_input("enter 80D deduction:")
    d4=st.number_input("enter 80G deduction:")
    d5=st.number_input("enter 80E educational  deduction:")
    d6=st.number_input("enter 80EEA housing loan deduction max upto1.5lakh:")
    d7=st.number_input("enter 80CCD employees contribution  deduction:")

    total_d=d1+d2+d3+d4+d5+d6+d7+standard
    st.write("total deductions:",total_d)
    ti=income-total_d




    if ti<=250000:
        st.write("no need  to pay any tax")
    elif(ti>250000 and ti<=500000):
        print("you have to pay 5% tax on incresed money")
        i_m=ti-250000
        tax_0=(5/100)
        st.write("you need to pay",i_m*tax_0)
    elif(ti>500000 and ti<=1000000):
        st.write("you have to pay 20% tax on incresed money")
        tax_2=(ti-500000)*(20/100)
        tax_1=(500000-250000)*(5/100)
        st.write("you need to pay ",tax_2+tax_1)
    elif(ti>1000000 ):
        st.write("you have to pay 20% tax on incresed money")
        tax_3=(ti-1000000)*(30/100)
        tax_2=(1000000-500000)*(20/100)
        tax_1=(500000-250000)*(5/100)
        st.write("you need to pay ",tax_3+tax_2+tax_1)


        
        


        

               
if regime=="new":
    taxable_income=income-standard
    
    #if st.button("submit"):
        #pass
       
    if taxable_income<300000:
        st.write('No need to pay any tax')

    if taxable_income>=300000 and taxable_income<600000:
        a=((taxable_income-300000)*0.05)
        st.write("you need to pay",((taxable_income-300000)*0.05)+(a*0.04),"with education tax i.e.4% on total tax")

    if taxable_income>=600000 and taxable_income<900000:
        b=(((taxable_income-600000)*0.1)+15000)
        st.write("you need to pay",(((taxable_income-600000)*0.1)+15000)+(b*0.04),"with education tax i.e.4% on total tax")

    if taxable_income>=900000 and taxable_income<1200000:
        c=(((taxable_income-900000)*0.15)+45000)
        st.write("you need to pay",(((taxable_income-900000)*0.15)+45000)+(c*0.04),"with education tax i.e.4% on total tax")

    if taxable_income>=1200000 and taxable_income<1500000:
        d=(((taxable_income-1200000)*0.2)+90000)
        st.write("you need to pay",(((taxable_income-1200000)*0.2)+90000)+(d*0.04),"with education tax i.e.4% on total tax")

    if taxable_income>=1500000:
        e=(150000+(taxable_income-1250000)*0.3)
        st.write("you need to pay",(150000+(taxable_income-1250000)*0.3)+(e*0.04),"with education tax i.e.4% on total tax")



















```
